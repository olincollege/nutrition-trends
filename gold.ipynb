```python
# Connecting to Google
from pytrends.request import TrendReq
import matplotlib as plt

pytrends = TrendReq(
    hl="en-US", tz=360
)  # hl means English #tz is something to do with time (Esther searches up later )
```

```python
# Building a payload which is what all of the variables abide to
kw_list = [
    "Apple",
    "Bread",
    "Chocolate",
    "Milk",
]  # the food names avaialbe from USDA (figure out if I get the kw_list to be found from the USDA thingie)
```

```python
pytrends.build_payload(
    kw_list,
    cat="71",
    timeframe="today 5-y",  # need to go in and change to 2008-present, as the web shopping can't fix (but could also just ignore)
    geo="US",
    gprop="froogle",  # specifically for the web shopping aspect
)  # specifiying when, country as US, category as
```

```python
# Interest over Time

# Getting Data
interest_data = pytrends.interest_over_time()
interest_data = interest_data.head(10)
print(interest_data)

# Graphing Data
interest_data.plot(figsize=(7, 5))
```
